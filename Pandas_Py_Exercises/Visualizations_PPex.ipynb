Pandas Visualization

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv(r"C:\Users\shubh\Downloads\Ice Cream Ratings.csv")
df = df.set_index('Date')
df
```

```python
print(plt.style.available)
plt.style.use('ggplot')
```

```python
df.plot() # Default = Line Graph
```

```python
df.plot(subplots=True)
```

```python
df.plot(title='Ice Cream Ratings', xlabel='Daily Ratings', ylabel='Scores')
```

```python
df.plot(kind='bar', stacked=True)
```

```python
df['Flavor Rating'].plot(kind='bar', stacked=True)
```

```python
df.plot.barh(stacked=True)
```

```python
df.plot.scatter(x='Texture Rating', y='Overall Rating', s=500, c='Pink')
```

```python
df.plot.hist(bins=20) # Bins default = 10
```

```python
df.boxplot()
```

```python
df.plot.area(figsize = (10,5))
```

```python
df.plot.pie(y='Flavor Rating', figsize = (10,6))
```
